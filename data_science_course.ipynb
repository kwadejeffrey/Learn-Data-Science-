```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd

from plotly.offline import init_notebook_mode, iplot, plot, download_plotlyjs

import cufflinks as cf
init_notebook_mode(connected=True)
cf.go_offline()

rx1 = np.linspace(0, 1, 100)
ry0 = np.random.randn(100) + 5
ry1 = np.random.randn(100) - 5
ry2 = np.random.randn(100)

fig = go.Figure()
fig.add_trace(go.Scatter(x=rx1, y=ry0, mode='lines', name='Rand 1'))
fig.add_trace(go.Scatter(x=rx1, y=ry1, mode='markers', name='Rand 2'))
fig.add_trace(go.Scatter(x=rx1, y=ry2, mode='lines+markers', name='Rand 3'))
```

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd

from plotly.offline import init_notebook_mode, iplot, plot, download_plotlyjs

import cufflinks as cf
init_notebook_mode(connected=True)
cf.go_offline()

# Adding Details to the plot

aapl_df = pd.read_csv("AAPL.csv")
x = aapl_df['Date']
y = aapl_df['Adj Close']
aapl_df;

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='AAPL'))
fig.update_xaxes(rangeslider_visible=True, title="Zoom in to see more")
fig.update_yaxes(title="AAPL Stock Price")
```
